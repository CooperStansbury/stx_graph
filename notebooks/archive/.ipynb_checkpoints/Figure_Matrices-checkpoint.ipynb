```python
import pandas as pd
import numpy as np
import os
import sys
from scipy.ndimage import gaussian_filter

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
nCells = 250
nGenes = 500
cdf = np.ones((nGenes, nCells))
noise = np.random.normal(0, 0.5, size=(nGenes, nCells))

cdf[0:100,0:50] = cdf[0:100,0:50] + np.random.normal(1, 1.1, size=cdf[0:100,0:50].shape)
cdf[100:,50:] = cdf[100:, 50:] + np.random.normal(1.1, 0.5, size=cdf[100:,50:].shape)
# cdf[400:,195:]  = cdf[400:,195:]  + np.random.normal(0.5, 0.9, size=cdf[400:,195:] .shape)
cdf = cdf + noise

cdf = gaussian_filter(cdf, sigma=1.2)


plt.rcParams['figure.dpi'] = 300
plt.imshow(cdf, cmap='viridis')
plt.axis(False)
```

```python
nSpots = 30

def f(x, y):
    return np.sin(0.3 * np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-1, 10, nSpots)
y = np.linspace(-25, 1, nSpots)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)


noise = np.random.normal(0, 0.3, size=Z.shape)
Z = Z - Z.mean()
cdf = Z + noise
# cdf = cdf + noise

cdf = gaussian_filter(cdf, sigma=1.2)


plt.rcParams['figure.dpi'] = 300
plt.imshow(cdf, cmap='viridis')
plt.axis(False)
```
