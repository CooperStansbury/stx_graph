```python
import pandas as pd
import numpy as np
import scipy
import sys
import os
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
from importlib import reload

# local imports 
import graph as gr
import utils as ut
```

```python
# load the data
```

```python
key = "ND"

dirpath = "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/graph_data/"
coordpath = f"{dirpath}coordinates.pq"
distpath = f"{dirpath}/distances/{key}_euclidean_distances.pq"
edgepath = f"{dirpath}/edge_lists/{key}_harmonic_highlevel_edgelist.pq"
stxpath = f"{dirpath}/gene_expression/{key}_spatial_cpm.pq"

cdf = pd.read_parquet(coordpath)
cdf = cdf[cdf['key'] == key].reset_index(drop=True)
print(f"{cdf.shape=}")
# print(cdf.head())

D = pd.read_parquet(distpath)
D = D.drop(columns='key')
print(f"{D.shape=}")
# print(D.head())

E = pd.read_parquet(edgepath)
print(f"{E.shape=}")
# print(E.head())

sdf = pd.read_parquet(stxpath, engine='pyarrow')
print(f"{sdf.shape=}")

# merge em!
df = pd.merge(D, E, 
              how='left',
              left_on=['node1', 'node2'],
              right_on=['node1', 'node2'])

print(f"{df.shape=}")
print(df.head())
```

# Build random neighborhoods

```python
reload(gr)

sample_size = 1
n = 500
r = 45

graphs = []

for i in range(sample_size):
    # point = cdf[['x', 'y']].sample(1).to_numpy().ravel()

    nbrhd = gr.get_neighborhood(cdf, 
                                center=True, 
                                n=n, 
                                metric='minkowski')

    # subset the edges and coords of the spaitial data
    coords = cdf[cdf['nodeId'].isin(nbrhd)].reset_index()
    edges = df[(df['node1'].isin(nbrhd)) & (df['node2'].isin(nbrhd))].reset_index()
    edges = edges[edges['d'] < r]
    stx = sdf[sdf.index.isin(nbrhd)]
    print(f"{coords.shape=} {edges.shape=} {stx.shape=}")

    corr = []
    pvals = []

    # # compute correlations
    for n1, n2 in edges[['node1', 'node2']].values:

        g1 = stx.loc[n1, :].values
        g2 = stx.loc[n2, :].values

        score, pval = scipy.stats.pearsonr(g1, g2)
        corr.append(score)
        pvals.append(pval)
    

    edges['correlation'] = corr
    edges['pvals'] = pvals
    
    G = gr.build_graph(edges, coords)
    graphs.append(G)


print('done')
```

```python
# plot the landscape

relationship = 'correlation'

plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.facecolor'] = "none"
plt.rcParams['figure.figsize'] = 9, 9

# plotting params
lut = 5
cmap = plt.cm.get_cmap('RdYlGn', lut=lut)
fig, ax = plt.subplots()

# node coloring
centrality = nx.eigenvector_centrality(G, weight=relationship, tol=0.001)
colors = [centrality[n] for n in G.nodes()]

# network plots
nx.draw_networkx_nodes(G,
                       pos=G.pos,
                       node_size=85,
                       # node_color=colors,
                       node_color='lightgrey',
                       # cmap=cmap,
                       # alpha=0.75,
                       edgecolors='k',
                       linewidths=1.5,
                       ax=ax)

eweights = np.array([e[relationship] for node1, node2, e in G.edges(data=True)])

nx.draw_networkx_edges(G,
                       pos=G.pos,
                       width=5,
                       # edge_color='k',
                       edge_color=eweights,
                       edge_cmap=cmap, 
                       edge_vmin=0, 
                       edge_vmax=1,
                       alpha=0.75,
                       ax=ax)   

# sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin = -1, vmax=1))
# sm._A = []
# plt.colorbar(sm)

ax.set_aspect('equal')
ax.axis(False)
```
