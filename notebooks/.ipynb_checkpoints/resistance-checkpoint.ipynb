```python
import pandas as pd
import numpy as np
import scipy
import sys
import os
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
from importlib import reload

# local imports 
import graph as gr
```

```python
# load the data
```

```python
key = "HFD8"

dirpath = "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/graph_data/"
coordpath = f"{dirpath}coordinates.pq"
distpath = f"{dirpath}/distances/{key}_euclidean_distances.pq"
edgepath = f"{dirpath}/edge_lists/{key}_harmonic_highlevel_edgelist.pq"

cdf = pd.read_parquet(coordpath)
cdf = cdf[cdf['key'] == key].reset_index(drop=True)
print(f"{cdf.shape=}")
print(cdf.head())

D = pd.read_parquet(distpath)
D = D.drop(columns='key')
print(f"{D.shape=}")
# print(D.head())

E = pd.read_parquet(edgepath)
print(f"{E.shape=}")
# print(E.head())

# merge em!
df = pd.merge(D, E, 
              how='left',
              left_on=['node1', 'node2'],
              right_on=['node1', 'node2'])

print(f"{df.shape=}")
print(df.head())
```

# Build random neighborhoods

```python
reload(gr)

sample_size = 1000
n = 100
t = 45

graphs = []

for i in range(sample_size):
    point = cdf[['x', 'y']].sample(1).to_numpy().ravel()
    nbrhd = gr.get_neighborhood(cdf, 
                                center=point, 
                                n=n, 
                                metric='minkowski')

    # define a graph
    G = gr.build_graph(df, cdf, nbrhd, t)

    graphs.append(G)

print('done')
```

# Compute Network Resistance

```python
q = 0.01
weight = 'innate_innate'
edgeThreshold = np.quantile(df[weight], q)
print(f"{weight}: {q=} {edgeThreshold=:.3f}")

G = graphs[0]
node_list = list(G)
A = nx.adjacency_matrix(G, node_list, weight=weight)
A = A.todense()      
Ahat = np.where(A > edgeThreshold, 1, 0)
print(f"{A.shape=} {A.min()=} {A.max()=:.3f}")

fig, ax = plt.subplots(1, 2)
ax[0].imshow(A)
ax[1].imshow(Ahat)

gr.effective_resistance(Ahat)
```

# Histogram of Resistances

```python
reload(gr)


weight = 'innate_innate'
# weight = 'adaptive_adaptive'
# q = 0.001
# edgeThreshold = np.quantile(df[weight], q)
# print(f"{weight}: {q=} {edgeThreshold=:.3f}")
edgeThreshold = 0


res = []

for i, G in enumerate(graphs):

    node_list = list(G)
    A = nx.adjacency_matrix(G, node_list, weight=weight)
    A = A.todense()      
    Ahat = np.where(A > edgeThreshold, 1, 0)
    # print(f"{A.shape=} {A.min()=} {A.max()=:.3f}")
    
    r = gr.effective_resistance(Ahat)

    row = {
        'graph' : i,
        'r' : r, 
        'n_edges' : np.sum(Ahat),
        'max_edge_weight' : A.max(),
        'min_edge_weight' : A.min(),
        'mean_edge_weight' : A.mean(),
        'median_edge_weight' : np.median(A),
    }
    res.append(row)


res = pd.DataFrame(res)
res.head()
```

# Nodal Resistance

```python
reload(gr)
G = graphs[0]
weight = 'innate_innate'


nodes = list(G)
Q = gr.laplacian_inverse(G, weight=weight)

# get two random nodes
n = Q.shape[0]
i, j = np.random.choice(range(n), 2, replace=False)
ni = nodes[i]
nj = nodes[j]
toplot = [ni, nj]


# get node vectors
ei = np.zeros(n)
ei[i] = 1

ej = np.zeros(n)
ej[j] = 1

r_ij = (ei - ej).T @ Q @ (ei - ej)
print(f"{r_ij=}")

t = nx.resistance_distance(G, ni, nj, weight=weight)
print(f"{t=}")


rd = gr.graph_resistance(G, weight=weight)
print(f"effective resistance: {rd=}")

# get the shortest path based on distance
path = nx.shortest_path(G, source=ni, target=nj, weight=None, method='dijkstra')
edgelist = []

for i in range(len(path)-1):
    tup = (path[i], path[i+1])
    edgelist.append(tup)

# network plots
plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.facecolor'] = "w"
plt.rcParams['figure.figsize'] = 4, 4

fig, ax = plt.subplots()

nx.draw_networkx_nodes(G,
                       pos=G.pos,
                       node_size=50,
                       node_color='lightgrey',
                       edgecolors='k',
                       linewidths=1,
                       ax=ax)

nx.draw_networkx_edges(G,
                       pos=G.pos,
                       width=6,
                       edge_color='w',
                       alpha=1,
                       ax=ax)   


# source and target nodes
nx.draw_networkx_nodes(G,
                       pos=G.pos,
                       node_size=50,
                       nodelist=toplot,
                       node_color='r',
                       edgecolors='k',
                       linewidths=1,
                       ax=ax)

nx.draw_networkx_edges(G,
                       pos=G.pos,
                       edgelist=edgelist,
                       width=3,
                       edge_color='r',
                       alpha=1,
                       ax=ax) 
```
