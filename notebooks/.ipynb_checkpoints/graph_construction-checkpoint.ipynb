```python
import pandas as pd
import numpy as np
import scipy
import sys
import os
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
from importlib import reload

# local imports 
import graph as gr
```

```python
# load the data
```

```python
key = "ND"

dirpath = "/nfs/turbo/umms-indikar/shared/projects/spatial_transcriptomics/graph_data/"
coordpath = f"{dirpath}coordinates.pq"
distpath = f"{dirpath}/distances/{key}_euclidean_distances.pq"
edgepath = f"{dirpath}/edge_lists/{key}_harmonic_highlevel_edgelist.pq"

cdf = pd.read_parquet(coordpath)
cdf = cdf[cdf['key'] == key].reset_index(drop=True)
print(f"{cdf.shape=}")
print(cdf.head())

D = pd.read_parquet(distpath)
D = D.drop(columns='key')
print(f"{D.shape=}")
# print(D.head())

E = pd.read_parquet(edgepath)
print(f"{E.shape=}")
# print(E.head())

# merge em!
df = pd.merge(D, E, 
              how='left',
              left_on=['node1', 'node2'],
              right_on=['node1', 'node2'])

print(f"{df.shape=}")
print(df.head())
```

# Network Construction

```python
# make the positions
pos = {}
for idx, row in cdf.iterrows():
    pos[row['nodeId']] = np.array(row[['x', 'y']])


threshold = 50
edges = df[df['d'] < threshold]

# make the graph object
G = nx.from_pandas_edgelist(edges, 
                            source='node1',
                            target='node2',
                            edge_attr=True)
G.pos = pos

print('done')
```

```python
relationship = 'innate_innate'

plt.rcParams['figure.dpi'] = 300
plt.rcParams['figure.facecolor'] = "none"
plt.rcParams['figure.figsize'] = 7, 7

# plotting params
lut = 5
cmap = plt.cm.get_cmap('Reds', lut=lut)
fig, ax = plt.subplots()

# node coloring
centrality = nx.eigenvector_centrality(G, weight=relationship, tol=0.001)
colors = [centrality[n] for n in G.nodes()]

# network plots
nx.draw_networkx_nodes(G,
                       pos=G.pos,
                       node_size=10,
                       node_color=colors,
                       cmap=cmap,
                       alpha=0.75,
                       edgecolors='k',
                       linewidths=0.25,
                       ax=ax)

eweights = np.array([e[relationship] for node1, node2, e in G.edges(data=True)])

nx.draw_networkx_edges(G,
                       pos=G.pos,
                       width=2,
                       edge_color=eweights,
                       edge_cmap=cmap, 
                       edge_vmin=0, 
                       edge_vmax=1,
                       alpha=0.25,
                       ax=ax)   

ax.set_aspect('equal')
ax.invert_yaxis()
# ax.axis('off')
```

```python

rels = [
    'innate_innate', 
    'innate_adaptive',
    'innate_other', 
    'adaptive_adaptive',
    'adaptive_other', 
    'other_other',
]

res = []

for rel in rels:
    nr, nrrl = nx.non_randomness(G, k=None, weight=rel)
    row = {
        'relation' : rel, 
        'nonrandomness' : nr,
        'rel_nonrandomness' : nrrl,}

    res.append(row)

res = pd.DataFrame(res)
res = pd.melt(res, id_vars='relation')
res.head()

```

```python
sns.barplot(data=res, 
            x='relation',
            y='value',
            hue='variable')

plt.gca().tick_params(axis='x', rotation=90)
```

```python

rels = [
    'innate_innate', 
    'innate_adaptive',
    'innate_other', 
    'adaptive_adaptive',
    'adaptive_other', 
    'other_other',
]

for rel in rels:
    A = nx.adjacency_matrix(G, weight=rel)
    A = A.todense()
    L = scipy.sparse.csgraph.laplacian(A, normed=True)
    
    lamba, vecs  = np.linalg.eigh(L)
    sns.lineplot(lamba, label=rel)


# plt.gca().set_yscale('log')
```

```python
# scores = nx.betweenness_centrality(G, weight=relationship)

scores = nx.fiedler_vector(G, 
                           weight=relationship, 
                           normalized=True, 
                           tol=1e-08, 
                           method='tracemin_pcg')


colors = [scores[n] for n in G.nodes()]

# plotting params
lut = 10
cmap = plt.cm.get_cmap('magma', lut=lut)
fig, ax = plt.subplots()

# network plots
nx.draw_networkx_nodes(G,
                       pos=G.pos,
                       node_size=15,
                       node_color=colors,
                       cmap=cmap,
                       alpha=1,
                       edgecolors='k',
                       linewidths=0.25,
                       ax=ax)


ax.set_aspect('equal')
ax.invert_yaxis()
```

```python
scores = nx.betweenness_centrality(G, weight=relationship)

colors = [scores[n] for n in G.nodes()]

# plotting params
lut = 10
cmap = plt.cm.get_cmap('magma', lut=lut)
fig, ax = plt.subplots()

# network plots
nx.draw_networkx_nodes(G,
                       pos=G.pos,
                       node_size=15,
                       node_color=colors,
                       cmap=cmap,
                       alpha=1,
                       edgecolors='k',
                       linewidths=0.25,
                       ax=ax)


ax.set_aspect('equal')
ax.invert_yaxis()
```

# Modularity

```python
resolution = 50
communities = nx.community.greedy_modularity_communities(G, 
                                                         weight=relationship,
                                                         resolution=resolution)

scores = pd.DataFrame(communities).T
scores = pd.melt(scores)
scores['variable'] += 1 
scores = pd.Series(scores['variable'].values,index=scores['value']).to_dict()

colors = [scores[n] for n in G.nodes()]


# plotting params
lut = resolution
cmap = plt.cm.get_cmap('magma', lut=lut)
fig, ax = plt.subplots()

# network plots
nx.draw_networkx_nodes(G,
                       pos=G.pos,
                       node_size=15,
                       node_color=colors,
                       cmap=cmap,
                       alpha=1,
                       edgecolors='k',
                       linewidths=0.25,
                       ax=ax)

ax.set_aspect('equal')
ax.invert_yaxis()
```
